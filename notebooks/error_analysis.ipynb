```python
def visualize_positions(agent_location: tuple[int, int], item_location: tuple[int, int], goal_location: tuple[int, int], fig: plt.Figure, ax: plt.Axes) -> None:
    n = 4
    ax.clear()
    ax.set_xlim(0, n)
    ax.set_ylim(0, n)
    ax.set_xticks(np.arange(0, n + 1, 1))
    ax.set_yticks(np.arange(0, n + 1, 1))
    ax.grid(True)

    # Plotting the agent, item, and goal
    ax.text(
        agent_location[0] + 0.5,
        agent_location[1] + 0.5,
        "A",
        ha="center",
        va="center",
        fontsize=16,
        color="blue",
    )
    ax.text(
        item_location[0] + 0.5,
        item_location[1] + 0.5,
        "I",
        ha="center",
        va="center",
        fontsize=16,
        color="green",
    )
    ax.text(
        goal_location[0] + 0.5,
        goal_location[1] + 0.5,
        "G",
        ha="center",
        va="center",
        fontsize=16,
        color="red",
    )

    handles = [
        plt.Line2D(
            [0], [0], marker="o", color="w", markerfacecolor="purple", markersize=8, label="Agent (A) with item"
        ),
        plt.Line2D([0], [0], marker="o", color="w", markerfacecolor="green", markersize=8, label="Item (I)"),
        plt.Line2D([0], [0], marker="o", color="w", markerfacecolor="red", markersize=8, label="Goal (G)"),
    ]
    ax.legend(handles=handles, loc="center left", bbox_to_anchor=(1, 0.5))

    plt.subplots_adjust(right=0.75, left=0.1)
    fig.canvas.draw_idle()
    plt.pause(2)
```

```python
failed_envs = [[list(map(int, position.split(", "))) for position in one_env.lstrip("(").rstrip(")").split(") (")] for one_env in raw_failed_envs.split("\n")]
```

From visualization, we can see that inifinite loop mostly occurs when agent is close to the item but needs to go to the goal, or when it is close to the goal but it should go to the item first.

```python
fig, ax = plt.subplots(figsize=(8, 8))
for failed_env in failed_envs:
    visualize_positions(*failed_env, fig, ax)
    break  # note: remove this line to visualize all failed environments
```

Numerically, we see that at least half of the failes samples have distance == 1 to either goal or item. 

```python
def are_neighbours(object_a_location: tuple[int, int], object_b_location: tuple[int, int]) -> bool:
    return abs(object_a_location[0] - object_b_location[0]) + abs(object_a_location[1] - object_b_location[1]) == 1
```

```python
samples_close_to_item = len([failed_env for failed_env in failed_envs if are_neighbours(failed_env[0], failed_env[1])])
samples_close_to_goal = len([failed_env for failed_env in failed_envs if are_neighbours(failed_env[0], failed_env[2])])
all_samples = len(failed_envs)

print(f"Samples close to item: {samples_close_to_item}/{all_samples}")
print(f"Samples close to goal: {samples_close_to_goal}/{all_samples}")
print(f"Samples close to item or goal: {samples_close_to_item + samples_close_to_goal}/{all_samples}")
print(f"Samples close to item and goal: {len([failed_env for failed_env in failed_envs if are_neighbours(failed_env[0], failed_env[1]) and are_neighbours(failed_env[0], failed_env[2])])}/{all_samples}")
```

Possible Solutions:
- not always stop the episode when it goes into an infinite loop but suggest another route when stack.
- give rewards dynamically depends on how close the agent is to the goal / item (i think this is not allowed)
- add revisiting penalty (which we had originally but removed because it increased the complexity of learning, but rather we should think about how to reduce this complexity with such rewards)
